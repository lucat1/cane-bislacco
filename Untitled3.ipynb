```python
import numpy as np
import tensorflow as tf
import keras
import numpy as np
from keras import layers
from keras import activations
from keras import losses
from sklearn.model_selection import train_test_split
import keras.backend as K

def polar_generator(batchsize,grid=(10,10),noise=.002,flat=False):
  while True:
    x = np.random.rand(batchsize)
    y = np.random.rand(batchsize)
    out = np.zeros((batchsize,grid[0],grid[1]))
    xc = (x*grid[0]).astype(int)
    yc = (y*grid[1]).astype(int)
    for b in range(batchsize):
      out[b,xc[b],yc[b]] = 1
    #compute rho and theta and add some noise
    rho = np.sqrt(x**2+y**2) + np.random.normal(scale=noise)
    theta = np.arctan(y/np.maximum(x,.00001)) + np.random.normal(scale=noise)
    if flat:
      out = np.reshape(out,(batchsize,grid[0]*grid[1]))
    yield ((theta,rho),out)
```

```python
size = 10
batch_size = 512
```

```python
n_train = 100000
n_test = 20000

def xy(i):
  (theta,rho),y = i
  # res = map(map_map, y)
  # y = np.array(list(res))
  #x=np.array([i for i in zip(theta,rho)])
  return ([theta,rho],y)

gcd = n_test
gen = polar_generator(gcd,grid=(size,size),noise=0.02,flat=True)

# x,y=xy(next(gen))
x_train, y_train = xy(next(gen))
for i in range(int(n_train/gcd)-1):
  x, y = xy(next(gen))
  x_trian = np.concatenate((x, x_train), axis=0)
  y_trian = np.concatenate((y, y_train), axis=0)

x_test, y_test = xy(next(gen))
for i in range(int(n_train/gcd)-1):
  x, y = xy(next(gen))
  x_trian = np.concatenate((x, x_test), axis=0)
  y_trian = np.concatenate((y, y_test), axis=0)


# x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=n_test/(n_train+n_test), shuffle=True, random_state=1)
```

```python
def my_categorical_accuracy(y_true: tf.Tensor, y_pred: tf.Tensor) -> float:
  equal = tf.equal(tf.math.argmax(y_true, 1), tf.math.argmax(y_pred , 1))
  n_correct = tf.reduce_sum(tf.cast(equal, tf.int32))
  return tf.cast(n_correct/len(y_true), tf.keras.backend.floatx())

def point_accuracy(y_true: tf.Tensor, y_pred: tf.Tensor) -> float:
  norm = tf.cast(tf.round(y_pred), tf.uint8)
  return tf.equal(tf.cast(y_true, tf.uint8), norm)
```

```python
input_theta = layers.Input(shape=(1,))
input_rho = layers.Input(shape=(1,))
dense_theta = layers.Dense(5, activation=activations.swish)(input_theta)
dense_theta_2 = layers.Dense(5, activation=activations.swish)(dense_theta)
dense_rho = layers.Dense(5, activation=activations.swish)(input_rho)
dense_rho_2 = layers.Dense(5, activation=activations.swish)(dense_rho)
concatenate_layer = layers.Concatenate()([dense_theta_2, dense_rho_2])
dense_layer = layers.Dense(5, activation=activations.swish)(concatenate_layer)
dense_layer_2 = layers.Dense(5, activation=activations.swish)(dense_layer)
output = layers.Dense(100, activation=activations.softmax)(dense_layer_2)
model = keras.Model([input_theta, input_rho], output)
model.build((None, 2))
model.summary()
model.compile(
    optimizer=keras.optimizers.Adam(learning_rate=1e-3),
    loss=losses.CategoricalCrossentropy(),
    metrics=['categorical_accuracy', my_categorical_accuracy]
)
```

```python
history = model.fit(
    [x_train],
    y_train,
    epochs=200,
    batch_size=batch_size
)
```

```python
score, _, acc = model.evaluate(x_test, y_test, batch_size=batch_size)
print('Test score:', score)
print('Accuracy: {:.1f}%'.format(acc*100))

accs = []
for x in range(50):
  x,y = xy(next(gen))

  score, _, acc = model.evaluate(x, y, batch_size=batch_size)
  accs.append(acc)

acc = np.mean(accs)
print('Accuracy: {:.1f}%'.format(acc*100))
```
